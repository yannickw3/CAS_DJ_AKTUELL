```python
## 3. Dictionary mit allen Inhalten (musste pdfminer weglassen, weil Fehlermeldung)


newlist = []

for file in filelist:
    byts = textract.process("pdfs/"+file, encoding="ascii")
    text = byts.decode("ascii").replace("\n", "") 
    
    mini_dict = {"Filename":file, "Text": text}
    
    newlist.append(mini_dict)
```

```python
# andere Möglichkeit, um Fehlermeldung zu umgehen:

newlist = []

for file in filelist:
    try:
        byts = textract.process("pdfs/"+file, encoding="ascii")
        text = byts.decode("ascii").replace("\n", "") 
        
        mini_dict = {"Filename":file, "Text": text}
        
        newlist.append(mini_dict)
    except:
        mini_dict = {"Filename":file, "Text": "Fehlermeldung"}
        newlist.append(mini_dict)
```

```python
# Command Line: pip install progressbar2

import progressbar

newlist = []
bar = progressbar.ProgressBar()

progresslist = range(len(filelist))

for file,b in zip(filelist, bar(progresslist)):
    byts = textract.process("pdfs/"+file, encoding="ascii")
    text = byts.decode("ascii").replace("\n", "") 
    
    mini_dict = {"Filename":file, "Text": text}
    
    newlist.append(mini_dict)
```

```python
## 4. Tabelle bauen mit Pandas

import pandas as pd

pd. DataFrame(newlist)
```

```python
# erkennt einige Files nicht als Text, sondern als Bilder > deshalb Methode tesseract

byts = textract.process("pdfs/Vk_20180514_49documents1.pdf", method="tesseract", encoding="ascii")
text = byts.decode("ascii").replace("\n", "")
text
```

```python
newlist = []
bar = progressbar.ProgressBar()

progresslist = range(len(filelist))

for file,b in zip(filelist, bar(progresslist)):
    byts = textract.process("pdfs/"+file, encoding="ascii")
    text = byts.decode("ascii").replace("\n", "") 
    
    if len(text) < 40:
        byts = textract.process("pdfs/"+file, method="tesseract", encoding="ascii")
        text = byts.decode("ascii").replace("\n", "")
    
    mini_dict = {"Filename":file, "Text": text}
    
    newlist.append(mini_dict)
```

```python
df = pd. DataFrame(newlist)
df
```

```python
# das Datum ist jeweils im Filename enthalten

test = "Vk_20180521_01documents2.pdf"
# string nach Underline aufsplitten und nur das zweite ELement ausgeben
test.split("_")[1]
```

```python
def date(elem):
    return elem.split("_")[1]

date(test)
```
